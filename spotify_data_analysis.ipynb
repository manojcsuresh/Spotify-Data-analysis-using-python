```python
corr_df=df_tracks.drop(["key", "mode", "explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot= True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linecolor="black", linewidths=1 )
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotaion=90)
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = df_tracks, x = 'energy', y= 'loudness', color = 'c')
plt.title('Energy vs Loudness')
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = df_tracks, x = 'acousticness', y= 'popularity', color = 'b')
plt.title('Acousticness vs Popularity')
plt.show()
```
